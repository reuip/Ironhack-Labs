## Prepare environment and import data

```python
#Preparing environment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing file into dataframe
df=pd.read_csv('data/Data_Marketing_Customer_Analysis_Round3.csv')
```

## Separate numeric and categorical data

```python
numericals=df.select_dtypes(np.number)
categoricals=df.select_dtypes(object)
```

## Check normality of data

### Seaborn distribution plots (used histplot instead)

```python
#Collect list of columns in order to create histogram for each one
x=numericals.columns

#For each column name, create a figure for it and create a histplot within that.  
#Use density feature since that appears to be the function of the depreciated distplot function

for i in x:
    plt.figure(figsize=(4,5))
    sns.histplot(x=i,data=numericals, stat='density') 
    
```

### Use Matplotlib to construct histograms

```python
numericals.hist(figsize=(15,20))
```

## Evaluate visually the histograms

None fully follow the normal distribution

Some values, such as number of open complaints, are discrete and therefore the histogram tells us less interesting information about it from a normalization perspective.

**Interesting are the skewed values -**
- Total claim amount being left skewed is probably good.  People tend to claim less from the insurance
- monthly premium is left skewed, meaning people tend to pay very little for their insurancee.  This could indicate little differentiation in the insurance offers
- customer lifetime value is also left skewed.  this aligns with the premiums and is probably a factor of the premium and the total claim amount.  again, this could be adjusted to be better in the future.
- number of policies tends to be closer to 1.  this means people do not buy supplemental insurances.

Months since policy inception is a flat line.  This can tell us that there is consistency in people signing up for policies.

Months since last claim seems to taper off at the right, showing that most people make claims within 2 years, and fewer have a claim only after 3 years.

### Understanding response by total claim amount - recoding variable

```python
sns.countplot(x=df['total_claim_amount'],hue=df['response'])
```

```python

fig,ax= plt.subplots(figsize = (10,5))
fig.canvas.draw()
df['tca_recoded']=pd.cut(df['total_claim_amount'],4)

ax=sns.countplot(x=df['tca_recoded'],hue=df['response']).set(title="Number of responses by total claim amount")

plt.xlabel("Claim amount brackets")
plt.legend(loc='upper right')
```

```python
#Because this gave weird tick marks, adjust with correct numbers by grabbing ranges using inline and min/max
min_tca=df['total_claim_amount'].min()
max_tca=df['total_claim_amount'].max()
range_tca=np.linspace(min_tca,max_tca,5)
range_tca
range_tca_tickmarks=[]
for i in range(4):
    range_tca_tickmarks.append(str(range_tca[i])+" to "+str(range_tca[i+1]))

fig,ax2= plt.subplots(figsize = (10,5))
df['tca_recoded']=pd.cut(df['total_claim_amount'],4)
ax2=sns.countplot(x=df['tca_recoded'],hue=df['response']).set(title="Number of responses by total claim amount",xticklabels=range_tca_tickmarks)
plt.xlabel("Claim amount brackets")
plt.legend(loc='upper right')
```

Show a plot of the response by income.


```python

fig,ax= plt.subplots(figsize = (10,5))
fig.canvas.draw()
df['income_recoded']=pd.cut(df['income'],4)

ax=sns.countplot(x=df['income_recoded'],hue=df['response']).set(title="Number of responses by income bracket")

plt.xlabel("Income brackets")
plt.legend(loc='upper right')
```
