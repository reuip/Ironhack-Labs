# Imports

```python
import sys
import pandas as pd
import requests
import qgrid


```

# Get relevant website and parse it. Check status

```python
url = 'https://www.billboard.com/charts/hot-100/'
response = requests.get(url)

from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser') 
response.status_code
```

# Exploratory: Find artist and songs

```python
soup.select('.o-chart-results-list__item .c-title')
```

```python
soup.select('.c-label.a-no-trucate')[0].get_text()
```

# For artist, temp store in string, strip \n,\t, then append to artist list

```python
artist_list=[]
for artist in range(100):
    artist_str=soup.select('.c-label.a-no-trucate')[artist].get_text()
    artist_str=artist_str.strip()
    artist_list.append(artist_str)
artist_list
```

# For song, do the same

```python
song_list=[]
for song in range(100):
    song_str=soup.select('.o-chart-results-list__item .c-title')[song].get_text()
    song_str=song_str.strip()
    song_list.append(song_str)
song_list
```

# Store results in a dataframe, export dataframe to csv

```python
top_100=pd.DataFrame(data={'song_name':song_list,'artist_name':artist_list})

new_index=list(range(1,101))
top_100['rank']=new_index
top_100.set_index('rank',inplace=True,drop=True)
top_100.to_csv(r'C:\Users\MichaelTaylo_c9zoof3\Documents\GitHub\Ironhack-Labs\Web scraping\top100.csv')
```
