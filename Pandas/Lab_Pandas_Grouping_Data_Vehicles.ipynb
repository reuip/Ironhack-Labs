## Convert impertial to metric

### CO2 Emissions

```python
cars['CO2 Emission Grams/Mile']=round(cars['CO2 Emission Grams/Mile']/1.60934,2)
cars.rename(columns={'CO2 Emission Grams/Mile': 'CO2 Emission Grams/KM'},inplace=True)

cars
```

### MPG --> KPL

```python
for i in ['City MPG','Highway MPG','Combined MPG']:
    cars[i]=round(cars[i]*(1.60934 / 3.78541),2)

imperial=['City MPG','Highway MPG','Combined MPG']
metric=['City KPL','Highway KPL','Combined KPL']
imp_to_metric_columns=dict(zip(imperial,metric))
cars.rename(columns=imp_to_metric_columns,inplace=True)
cars
```

Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Grams/Mile * Mile/Km -> Grams/Mile * 1 Mile/1.60934Km

$$ \frac{Grams}{Mile} * \frac{Mile}{Km} $$

$$ \frac{Grams}{Mile} * \frac{1 Mile}{1.60934Km}  $$

convert MPG columns to km_per_liter

MPG = Miles/Gallon -> Km/Liter

1 Mile = 1.60934 Km

1 Gallon = 3.78541 Liters

$$ \frac{Miles}{Gallon} -> \frac{Miles}{Gallon} * \frac{Km}{Miles} * \frac{Gallon}{Liters}$$

$$ \frac{Miles}{Gallon} -> \frac{Miles}{Gallon} * \frac{1.60934Km}{ 1Miles} * \frac{1 Gallon}{3.78541 Liters}$$

* ( 1.60934 / 3.78541 )


What brand has the most cars?

```python
#Do it without groupby
x=cars['Make'].value_counts()
print(x.index[0],'has the most cars with',x[0],'models')
```

```python
#Do it with groupby
cars.groupby('Make')[['Make']].value_counts().sort_values(ascending=False).head(1)
```

<b>show the average CO2_Emission_Grams/Km  by Brand

<b>show the average CO2_Emission_Grams/Km  by Brand ... sorted

# (Optional) 

Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars['Year'].describe()
```

```python
decade_list=[1980,1990,2000,2010,2020]
labels_list=['80s','90s','2000s','2010s']
cars['Decade']=pd.cut(cars['Year'],bins=decade_list,labels=labels_list)
```

```python
cars
```

### Did cars consume more gas in the eighties?

show the average City_Km/Liter by year_range

```python
cars.groupby("Decade")[['City KPL','Highway KPL','Combined KPL']].mean()
```

```python
cars.groupby('Decade').agg({'City KPL':'mean','Highway KPL':'mean','Combined KPL':'mean'})
```

Which brands are more environment friendly?

```python
cars.groupby(['Make','Decade'],observed=True)[['CO2 Emission Grams/KM']].mean().sort_values(by="CO2 Emission Grams/KM")
```

Does the drivetrain affect fuel consumption?

```python
cars.groupby(['Drivetrain']).agg({'Highway KPL':'mean','City KPL':'mean'})
```

Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
cars['Transmission'].unique()
```

```python
def update_trans(x):
    if x.find("Auto")!=-1:
        return "Automatic"
    elif x.find("Manual")!=-1:
        return "Manual"
    else:
        return x

cars['Transmission']=list(map(update_trans,cars['Transmission']))
cars['Transmission'].unique()
```

```python
cars.groupby("Transmission")[['City KPL']].mean()
```

Use `groupby` and `agg` with different aggregation measures for different columns:

aggregate with average City_Km/Liter and the count of the Trans

```python
cars.groupby("Transmission").agg({'City KPL':'mean','Transmission':'count'})
```

aggregate with average City_Km/Liter and the minimum of the Trans

```python
cars.groupby("Transmission").agg({'City KPL':['mean','min']})
```
